```python
import numpy as np
import pandas as pd
```

```python
# Criar dataframe a portir de um array

arr = np.random.randint(10, 99, size = (5,5))
arr
```

```python
df1 = pd.DataFrame(data = arr, columns = ['Brasil', 'Chile', 'Argentina', 'Peru', 'Equador'])
df1
```

```python
# Criar dataframe a partir de listas

lista = [[23, 25, 40, 22], [17, 74, 33, 67], [54, 18, 88, 67], [41, 12, 96, 46]]
df2 = pd.DataFrame(data = lista, index = ['A', 'B', 'C', 'D'] ,columns = ['Brasil', 'Chile', 'Argentina', 'Peru'])
df2
```

```python
# Criar dataframe a partir de dicionários

dic = {'Produto':['TV', 'Som', 'Game', 'Smart'], 'Preço':[1700, 704, 1933, 2967], 'Validade':[2021, 2022, 2022, 2022]}
df3 = pd.DataFrame(data = dic)
df3
```

```python
# Criando uma nova coluna

df3['Custo'] = [1500, 500, 1400, 2000]
df3
```

```python
# Cálculo aritmético entre colunas, já criando uma nova

df3['Lucro'] = df3['Preço'] - df3['Custo']
df3
```
