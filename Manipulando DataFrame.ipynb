```python
import numpy as np
import pandas as pd
np.random.seed(100)
```

```python
ind = ['Janeiro', 'Março', 'Maio', 'Julho', 'Agosto', 'Outubro', 'Dezembro']
col = np.arange(1, 32)
dados = np.random.randint(492, 2050, size = (7, 31))
```

```python
df = pd.DataFrame(data = dados, index = ind, columns = col)
df
```

#### Para encontrar algum dado específico em um dataframe

`df[coluna]`

`df[[col1, col2, col3]]`

`df[coluna][linha]`

`df[linha:linha]` se tiver os dois pontos, trabalhamos com as linhas

`df.loc[indice linha, indice coluna]`, por exemplo, `df.loc['Julho',30]` para encontrar o dado 2001

`df.iloc[ind num linha, ind num coluna]`, por exemplo, `df.iloc[3,30]`

`df.loc[[ind1, ind2, ind3],[col1, col2, col3]]`

```python
# Quando vendeu no dia 10 de todos os meses?

df[10]
```

```python
# Quando vendeu nos diad 10, 20 e 30 de todos os meses?

df[[10, 20, 30]]
```

```python
# Quando vendeu no dia 15 de março?

df[15]['Março']
```

```python
# Quando vendeu nos meses de Janeiro, Março e Maio?

df[0:3] # 0 1 2 -> 0:3
```

```python
# Quanto vendeu nos meses de julho, agosto e outubro nos dias 25 a 31?

df.loc[['Julho','Agosto', 'Outubro'],[25, 26,27, 28, 29, 30, 31]]
```

```python
df.iloc[[3,4,5],[24, 25, 26, 27, 28, 29, 30]]
```
